```python
from agent_lab.model_hub import LLM_THINKING
from langchain.chat_models import init_chat_model

deepseek_v4_flash_thinking = init_chat_model(**LLM_THINKING)
prompt = 'Why do parrots have colorful feathers?'
```

```python
complete_response = deepseek_v4_flash_thinking.invoke(prompt)
```

```python
for block in complete_response.content_blocks:
    print(block)
```

```python
complete_response.pretty_print()
```
